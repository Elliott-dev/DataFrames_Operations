```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
file_path = os.path.join('../','Resources', 'team-1.csv')
team_1_df = pd.read_csv(file_path)
```

```python
# Read the CSV file into a DataFrame
team_1_df.head()
#team_1_df.info()
#team_1_df.describe()
```

```python
# Plot the DataFrame before the uisng melt()
plot = team_1_df.plot.bar()
plot
```

```python
# Melt the DataFrame
team_1_df.melt()
```

```python
# Melt the DataFrame using the runner name.
team_1_df.melt(id_vars="runner_name")
```

```python
# Melt the DataFrame using the coach name.
team_1_df.melt(id_vars="coach_name")
```

```python
# Remove coach name from original DataFrame
team_1_df=team_1_df[['runner_name','run_1','run_2','run_3','run_4']]
team_1_df
```

```python
# Melt the updated original DataFrame using the runner_name
# column, and add a var_name parameter and a value_name 
# parameter to rename the columns.

renamed_team_1_df = team_1_df.melt(id_vars="runner_name", var_name="run", value_name="Distance")
renamed_team_1_df
```

```python
# Plot a bar chart to show how the data has changed after melting it.
plot = renamed_team_1_df.plot.bar()
```

```python
# Save melted DataFreame to a CSV file
renamed_team_1_df.to_csv (r'C:\Users\ellio\OneDrive\Desktop\Netflix Project\NFLX-VIRT-DATA-F-PT-01-2022-E-LOLC\02-Assignments\04-DataFrame_Operations\Resources\Melted_Dataframe.csv', header=True)
```

```python
# Convert the melted DataFrame from long form back to wide form using 
# the pivot_table() function.
renamed_team_1_df.pivot_table(values='Distance', index='runner_name', columns='run')
```
